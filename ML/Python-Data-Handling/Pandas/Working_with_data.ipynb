Loading Data from files

```python
import pandas as pd
```

pd.read_csv can be changed for different files types. E.g. Parquet, excel

```python
coffee = pd.read_csv('./warmup-data/coffee.csv')
results = pd.read_parquet('./data/results.parquet')
bios = pd.read_csv('./data/bios.csv')
```

Acessing data from files (Head, Tail), Sample() = Random data

```python
coffee.head(1)
coffee.tail(2)
coffee.sample(5)
```

.Loc = dataframe.loc[#rows, #columns] (you can slice and then column name)

```python
coffee.loc[0:3,["Day", "Coffee Type"]]
```

Iloc is the same as Loc but using Indexes

```python
coffee.iloc[0:3,[0, 1]]
```

Sorting values is sort_Values, you can sort by columns if you want, for example I did coffee_type then units_sold which first sorted expresso then Latte Units sold.

```python
coffee.sort_values("Units Sold", ascending=False)
coffee.sort_values(["Coffee Type", "Units Sold"], ascending=True)
```

Filtering bios with loc based on if the height > 220 and country birth

```python
bios.loc[(bios["height_cm"]>220) & (bios["born_country"] == "USA")]
```

Filtering based of if the name contains "Keith"

```python
bios.loc[bios["name"].str.contains("Keith")]
```
